```python
import mlflow
from synapse.ml.predict import MLFlowTransformer

df = spark.sql("select * from predict_features")
```

```python
model = MLFlowTransformer(
    inputCols=["a","b","c","d","e","f"],
    outputCol="y_p",
    modelName="experiment-01-best",
    modelVersion="1/lgbm",
)
```

```python
df = model.transform(df)
```

```python
display(df)
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("predicted_features")
```
