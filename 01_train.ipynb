```python
from flaml import AutoML
import mlflow

mlflow.set_experiment("experiment-01")

automl: AutoML = AutoML()
automl_settings: dict = {
    "time_budget": 300, # Time limit in seconds
    "task": "classification", # Type of machine learning task
    "log_file_name": "test_automl.log", # Name of the log file
    "metric": "log_loss", # Evaluation metric
    "log_type": "all", # Level of logging
}
```

```python
df = spark.sql("SELECT * FROM dbo.train_features").toPandas()


with mlflow.start_run(nested=True, run_name="experiment-01-trials"):
    automl.fit(dataframe=df, label="y", **automl_settings)
```

```python
df_predict = spark.sql("select * from dbo.predict_features")
```

```python
result = automl.predict_proba(df_predict.toPandas())
```

```python
df_predicted = df_predict.toPandas().assign(
    y_p=result[:,1]
)
```

```python
spark.createDataFrame(df_predicted).write.mode("overwrite").format("delta").saveAsTable("predicted_features")
```
